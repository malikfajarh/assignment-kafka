```python
import json
import uuid
import os
import json
from dotenv import load_dotenv
from pathlib import Path
from kafka import KafkaProducer
from faker import Faker
from time import sleep
import spotify_pb2 
```

```python
dotenv_path = Path('/resources/.env')
load_dotenv(dotenv_path=dotenv_path)
```

```python
kafka_host = os.getenv('KAFKA_HOST')
kafka_topic = os.getenv('KAFKA_TOPIC_NAME')
kafka_topic_partition = os.getenv('KAFKA_TOPIC_NAME')
```

```python
print(kafka_host)
```

```python
producer = KafkaProducer(
    bootstrap_servers=f'{kafka_host}:9092',
    # value_serializer=lambda v: v.SerializeToString()  # Serialize Protobuf
)
```

```python
_instance = Faker()
global faker
faker = Faker()
```

```python
class DataGenerator(object):
    @staticmethod
    def get_data():
        return [
                uuid.uuid4().__str__(),
                faker.word(),
                faker.random_element(elements=("The Weekend", "SZA", "Taylor Swift", "Coldplay", "John Mayer", "Maroon 5",
                                             "Justin Bieber", "Ed Sheeran", "Lady Gaga", "Drake", "Eminem",
                                              "David Guetta")),
                faker.random_element(elements=("Pop", "Rock", "Jazz", "Country", "Hip hop", "Soul", "R&B", "EDM")),
                faker.random_int(min=21, max=97),
                faker.random_int(min=2, max=5),
                faker.random_int(min=0, max= 178),
                faker.boolean(),
                faker.random_element(elements=("Play", "Stop", "Skip", "Like", "Dislike")),
                faker.random_element(elements=("United States", "Indonesia", "Spain", "England", "Brazil", "Italy",
                                              "Germany", "Portugal", "Japan", "Singapore", "Thailand", "India")),
                faker.unix_time()
        ]

for i in range(1,400):
    columns =  ["track_id", "track_name", "artist", "genre", "popularity", "Duration", "tempo", "explicit", "interaction", "listened_from", "ts"]
    
    data_list = DataGenerator.get_data()
    json_data = dict(
        zip(columns,data_list)
    )
    _payload = json.dumps(json_data).encode("utf-8")
    response = producer.send(topic=kafka_topic_partition, value=_payload)
    # response = producer.send(topic=kafka_topic_partition, value=_payload, partition=2)
    print(json_data['track_id'],response.get())
    print('-' * 100)
    # print(_payload)
    sleep(5)  # every 5 second generate random event
```
